Line Plot

```python
# impoerting the
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt

import seaborn as sns


file_data = pd.read_csv("fifa.csv",index_col="Date",parse_dates=True)
print(file_data.head())

plt.figure(figsize=(16,6))
plt.title("FIFA Ranking")
sns.lineplot(data = file_data)
plt.show()
```

Now making a subset of graphs

```python
file_data = pd.read_csv("fifa.csv",index_col="Date",parse_dates=True)

plt.figure(figsize=(16,6))
plt.title("FIFA Ranking")
sns.lineplot(data = file_data["ARG"],label ="ARG")# we just place the name of the single column instead of whole data
sns.lineplot(data = file_data["ESP"],label ="ARG")
plt.show()
```

```python
file_data.tail()
```

Bar Chart 

```python
plt.title("BAR PLOT")
sns.barplot(x = file_data.index,y = file_data['ARG'])
plt.show()
```

Heat Map

```python
sns.heatmap(data=file_data, annot=True) #This ensures that the values for each cell appear on the chart. (Leaving this out removes the numbers from each of the cells!)
plt.title("HEAT MAP")
plt.show()
```
