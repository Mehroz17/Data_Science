```python
title = []
links_of_the_add  = []
for i in link_of_pages:
    r = requests.get(i).text
    s = bs(r)
    title.append([a.get("title")for a in s.find_all('a',class_ = "product-name")])
    links_of_the_add.append([a.get("href")for a in s.find_all('a',class_ = "product-name")])
print(len(title))
print(len(links_of_the_add))
```
