<h1>Mobiles Phone Images 

```python
driver =webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://obiwezy.com/mobile-devices/phones.html")
```

clicking on load more until no more items


```python
while True:
    try:
        load =  driver.find_element(By.CLASS_NAME,"load-more")
        load.click()
    except:
        break
```

Going to the end of the page

```python

body = driver.find_element(By.TAG_NAME,'body')
body.send_keys(Keys.END)
```

```python
html = driver.page_source
```

```python
s = bs(html)
```

```python
il =[img.get("src") for img in  s.find_all('img', class_='product-image-photo lazyload main-img')]
len(il)
```

```python
for i in range(len(il)):
    with open(str(i)+".jpg",'wb') as f:
        im = requests.get(il[i])
        f.write(im.content)
```

<h1>Mobile phone images download form the given website

<h1>Now getting phones images from amazon

```python
# making links of amazon with 20pages
i = 1
links_of_amazon_mobiles = []
while(i<41):
    links_of_amazon_mobiles.append("https://www.amazon.co.uk/s?k=mobile+phones&page=3&qid=1678903548&sprefix=mobile%2Caps%2C265&ref=sr_pg_"+str(i))
    i = i+1
links_of_amazon_mobiles
```

Now getting links of the images

```python
# now getting the links of the pic from all the 20 pages of amazon
m1 = []
HEADERS = ({'User-Agent':
	            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
	            'Accept-Language': 'en-US, en;q=0.5'})
for i in links_of_amazon_mobiles:
        re = requests.get(i, headers= HEADERS).text
        s  = bs(re,'lxml')
        m = [n.get("src") for n in s.find_all("img",class_ = "s-image")]
        m1 = m1+m
```

```python
print(len(m1),"\n")
m1
```

```python
# now saving the images into the folder 
for i in range(0,800):
    with open("amazon"+str(i)+".jpg",'wb') as f:
        im = requests.get(m1[i])
        f.write(im.content)
```

<h1>Now it is for tablets

```python
driver2 =webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver2.get("https://obiwezy.com/mobile-devices/tablets.html")
```

This is the new script to go to the end of the page

```python

while True:
    try:
        # wait for the "Load More" button to appear and click it
        wait = WebDriverWait(driver2, 10)
        load_more_button = wait.until(EC.visibility_of_element_located((By.XPATH, "//button[contains(text(), 'Load More')]")))
        load_more_button.click()
        time.sleep(10)
    except:
        # stop the loop when there are no more items to load
        break

# do whatever you want with the loaded items
items = driver2.find_elements(By.XPATH,"//div[@class='product-item-info']")
print(f"Number of loaded items: {len(items)}")
```

```python
html2 = driver2.page_source
```

```python
bs2 = bs(html2)
bs2
```

```python
tablet_links =[img.get("src") for img in  bs2.find_all('img', class_='product-image-photo lazyload main-img')]
len(tablet_links)
tablet_links
```
